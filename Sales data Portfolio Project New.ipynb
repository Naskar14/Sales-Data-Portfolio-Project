# Question 1 - What was the best month for sale? and the total earning

```python
#in order to get the sales, first need to get the sales value by multiplying the order quantity with price
```

```python
df.info()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df.head()
```

```python
df['sales_value'] = (df['Quantity Ordered']) * (df['Price Each'])
df.head()
```

```python
month_grp = df.groupby('Month').sum()
month_grp['sales_value']
```

```python
import matplotlib.pyplot as pl 

months = range(1,13)

pl.bar(months, month_grp['sales_value'])
pl.title('Sales by Monht')
pl.xticks(months)
pl.xlabel('Month')
pl.ylabel('Sales Value')
pl.grid(True)
pl.show()
```

# Question no 2 - Which city has the highest no of sales. Result in city level

```python
#first need to get the city from the purchase address
```

```python
df.head()
```

```python
def get_city(address):
    return address.split(',')[1]

df['City'] = df['Purchase Address'].apply(lambda x: get_city(x))

df.head()
```

```python
# need to add the state alias as well along with the city name just to make sure we get unique city names
```

```python
# droping the city column as of now
df = df.drop(columns='City')
df
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))

df.head()
```

```python
city_grp = df.groupby('City').sum()
city_grp['sales_value']
```

```python
cities = [city for city, df in df.groupby('City')]

pl.bar(cities, city_grp['sales_value'])
pl.title('Sales by City')
pl.xticks(cities, rotation='vertical', size=8)
pl.xlabel('City names')
pl.ylabel('Sales value')
pl.show()
```

```python
from itertools import combinations
from collections import Counter 

count = Counter()

for row in df2['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key, value in count.most_common(10):
    print(key,value)
```
